```python
import pandas as pd
import numpy as np
```

```python
df_train = pd.read_csv("santander-product-recommendation/train_ver2.csv")
df_train.head()
```

```python
num_cols = [col for col in df_train.columns[:24] if df_train[col].dtype in ["int64","float64"]]
df_train[num_cols].head()
```

```python
cat_cols = [col for col in df_train.columns[:24] if df_train[col].dtype in ["O"]]
df_train[cat_cols].head()
```

```python
df_train[cat_cols].describe()
```

```python
# カテゴリ変数の固有値確認
for col in cat_cols:
    uniq = np.unique(df_train[col].astype(str))
    print("-"*50)
    print(f"col:{col}, length:{len(uniq)}, {uniq}")
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
skip_cols = ["ncodpers","renta"]
for col in df_train.columns:
    if col in skip_cols: continue

    print("-"*50)
    print(f"col:{col}")

    f,ax = plt.subplots(figsize=(20,15))
    sns.countplot(x=col,data=df_train,alpha=0.5)
    plt.show()
```
